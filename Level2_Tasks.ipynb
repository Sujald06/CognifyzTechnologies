```python
#Importing All Necessary Libraries
import pandas as pd
import folium
from IPython.display import display
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import geopandas
from geodatasets import get_path
from folium.plugins import HeatMap
```

```python
df = pd.read_csv("Dataset .csv")  #Loading the Dataset.csv
df
```

```python
print(df[["Longitude", "Latitude"]])
```

```python
gdf = geopandas.GeoDataFrame(

df, geometry = geopandas.points_from_xy(df.Longitude, df. Latitude), crs = "EPSG:4326"

)
```

```python
print(gdf.head())
```

```python
world = geopandas.read_file(get_path("naturalearth.land"))
ax = world.plot(color="white", edgecolor = "k")
gdf.plot(ax = ax, marker = 'o', color='lime', markersize = 5)
plt.show()
```

```python
world_Map =  folium.Map(location = [df['Latitude'].mean(), df['Longitude'].mean()], zoom_start = 1, height = '100%', width = '100%')
```

```python
Heat_Marker = [[row['Latitude'], row['Longitude']] for i, row in df.iterrows()]
HeatMap(Heat_Marker, radius = 10).add_to(world_Map)
world_Map
```

```python
# Group by 'Restaurant Name' and count the number of outlets
restaurant_chain = df.groupby("Restaurant Name").size().reset_index(name="OutletCount")

# Filter for restaurants with more than one outlet
filtered_chains = restaurant_chain[restaurant_chain["OutletCount"] > 1]

# Sort the results by 'OutletCount' in descending order
restaurant_chains = filtered_chains.sort_values(by="OutletCount", ascending=False)
```

```python
print(restaurant_chains [["Restaurant Name", "OutletCount"]].head(10))
```

```python
Top_5_chains = restaurant_chains[:5]
# Create a bar plot for the top 5 restaurant chains
plt.bar(Top_5_chains["Restaurant Name"], Top_5_chains["OutletCount"], color='cyan', edgecolor='k')
# Add labels and title
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 5 Restaurant Chains by Number of Outlets")
plt.show()
```

```python
# Calculate the mean aggregate rating for each restaurant
ratings = (df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating").sort_values(by="Average Rating", ascending=False))
print(ratings)
```

```python
votes = (df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes").sort_values(by="Total Votes", ascending=False))
print(votes)
```

```python
top_5_ratings = ratings.head(5)
colors = ['turquoise', 'teal', 'aqua', 'darkgreen', 'darkorange']
plt.figure(figsize=(10, 6))
plt.bar(x=top_5_ratings["Restaurant Name"], height=top_5_ratings["Average Rating"],color=colors,edgecolor='k')
plt.xlabel("Restaurant Chain", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.title("Top 5 Restaurant Chains by Average Rating", fontsize=14)
plt.tight_layout()
plt.show()
```
