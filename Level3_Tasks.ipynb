```python
#Importing All Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")  #Loading the Dataset.csv
df
```

```python
print(df['Rating text'])
```

```python
positive_data = df.loc[df['Aggregate rating'] >= 2.5]
print(positive_data)
```

```python
positive_keywords = positive_data['Rating text'].unique()
print(positive_keywords)
```

```python
from collections import Counter

keyword_counts = Counter()
for text in positive_data['Rating text']:
    for keyword in positive_keywords:
        if keyword in text:
            keyword_counts[keyword] += 1

# Extract keywords and their frequencies
keywords = list(keyword_counts.keys())
frequencies = list(keyword_counts.values())

# Plot the histogram (bar chart in this case)
plt.bar(keywords, frequencies, color='indigo', edgecolor='k')
plt.xlabel("Positive Keywords")
plt.ylabel("Frequency")
plt.title("Frequency of Positive Keywords in Ratings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

negative_data = df.loc[(df['Aggregate rating'] < 2.5) & (df['Aggregate rating'] > 0)]
#print(negative_data)
negative_keywords = negative_data['Rating text'].unique()
print(negative_keywords)
```

```python
plt.hist(negative_data['Rating text'], label = negative_keywords)
plt.show()
```

```python

Highest_voted_resturant = df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending = False)
Highest_voted_resturant = Highest_voted_resturant.idxmax()
print("The Highest Voted Resturent is : ",Highest_voted_resturant)
```

```python

Lowest_voted_restaurant = df.groupby('Restaurant Name')['Votes'].sum().sort_values()
Lowest_voted_restaurant = Lowest_voted_restaurant.idxmin()
print("The Lowest Voted Restaurant is : ",Lowest_voted_restaurant)
```

```python
import numpy as np

correlation = np.corrcoef(df['Votes'], df['Aggregate rating'])
correlation = np.round(correlation, 2)
print(correlation)
```

```python
correlation_matrix = np.corrcoef(df['Votes'], df['Aggregate rating'])
plt.figure(figsize=(10, 5), dpi=100)
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='cool', linewidths=0.4)
plt.title("Correlation Heatmap: Votes vs Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Aggregate rating', y='Votes', hue='Aggregate rating', palette='copper', edgecolor='k')
sns.regplot(data=df, x='Aggregate rating', y='Votes', scatter=False, color='blue', line_kws={'color': 'lime', 'lw': 2})
plt.title("Correlation between Aggregate Rating and Votes: 0.31", fontsize=14)
plt.tight_layout()
plt.show()
print("As per the result, the correlation value is 0.31, which indicates a moderate positive correlation between these two variables.")

```

```python
x1 = df['Has Online delivery']
y = df['Average Cost for two']
x2 = df['Has Table booking']
```

```python
plt.figure(figsize=(12, 5))
# Subplot 1: Online delivery vs Average Cost for two
plt.subplot(1, 2, 1)
sns.boxplot(x=x1, y=y, palette='pastel')
plt.xlabel('Online Delivery', fontsize=12)
plt.ylabel('Average Cost for Two', fontsize=12)
plt.title('Impact of Online Delivery on Cost', fontsize=14)
# Subplot 2: Table booking vs Average Cost for two
plt.subplot(1, 2, 2)
sns.boxplot(x=x2, y=y, palette='spring')
plt.xlabel('Table Booking', fontsize=12)
plt.ylabel('Average Cost for Two', fontsize=12)
plt.title('Impact of Table Booking on Cost', fontsize=14)
plt.tight_layout()
plt.show()
```
